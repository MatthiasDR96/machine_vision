# Thresholding

```python
# Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image
image = cv2.imread('../../data/sudoku.jpg')

# Convert BGR-space to RGB-space (OpenCV loads images in BGR-space)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(image)  
plt.show()  
```

```python
# Convert RGB-image to grey scale image
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Show image
plt.imshow(gray, cmap='gray') 
plt.show()  
```

```python
# Binary thresholding
ret, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Adaptive thresholding
thresh2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 5) 

# Otsu's thresholding
ret, thresh3 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
# Show images
plt.subplot(1, 4, 1)
plt.title("Original")
plt.imshow(gray, cmap='gray')
plt.subplot(1, 4, 2)
plt.title("Binary")
plt.imshow(thresh1, cmap='gray')
plt.subplot(1, 4, 3)
plt.title("Adaptive")
plt.imshow(thresh2, cmap='gray')
plt.subplot(1, 4, 4)
plt.title("Otsu")
plt.imshow(thresh2, cmap='gray')
plt.show()
```
