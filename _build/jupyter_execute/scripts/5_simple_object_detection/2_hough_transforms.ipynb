# Hough transforms

```python
# Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Hough line detection

```python
# Load images
image = cv2.imread('../../data/sudoku.jpg')

# Converting RGB image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Show image
plt.imshow(image)
plt.show()
```

```python
# Calculate the edges
canny = cv2.Canny(gray, 50, 200)

# Show image
plt.imshow(canny, cmap='gray')
plt.show()
```

```python
# Search Houghlines
lines = cv2.HoughLines(canny, 1, np.pi/180, 200)

# Run for loop through each line
for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x_1 = int(x0 + 1000 * (-b))
    y_1 = int(y0 + 1000 * (a))
    x_2 = int(x0 - 1000 * (-b))
    y_2 = int(y0 - 1000 * (a))
    cv2.line(image, (x_1, y_1), (x_2, y_2), (255, 0, 0), 2)
    
# Show Final output
plt.imshow(image)
plt.show()
```

## Hough circle detection

```python
# Load images
image = cv2.imread('../../data/beer_cans.jpg')

# Converting RGB image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Show image
plt.imshow(image)
plt.show()
```

```python
# Calculate the edges
canny = cv2.Canny(gray, 50, 200)

# Show image
plt.imshow(canny, cmap='gray')
plt.show()
```

```python
# Search Houghlines
rows = gray.shape[0]
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, rows / 8, param1=100, param2=30, minRadius=120, maxRadius=150)

# Run for loop through each line
for circle in circles[0, :]:
    center = (int(circle[0]), int(circle[1]))
    cv2.circle(image, center, 1, (0, 100, 100), 3)
    radius = int(circle[2])
    cv2.circle(image, center, radius, (255, 0, 255), 3)
    
# Show Final output
plt.imshow(image)
plt.show()
```
