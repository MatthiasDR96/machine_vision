# Simple classifier

```python
# Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Images
image = cv2.imread('../../data/square.png')
plt.imshow(image)
plt.show()
```

## Feature extraction

```python
# Converting RGB image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Compute edges
canny = cv2.Canny(gray, 50, 100)
plt.imshow(canny)

# Search Houghlines
lines = cv2.HoughLines(canny, 10, np.pi*10/180, 100)
print(lines)
```

## Classification

```python
# Classifier
if len(lines) == 4:
    print("Square")
elif len(lines) == 3:
    print("Triangle")
else:
    print("Undefined")
```
