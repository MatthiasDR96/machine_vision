## Template matching

```python
# Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load images
image = cv2.imread('../../data/mario.jpg')
template = cv2.imread('../../data/mario_coin.jpg', 0)

# Converting RGB image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Show image
plt.imshow(image)
plt.show()
```

```python
# Get shape of the template
w, h = template.shape[::-1]

# Template matching 
res = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF_NORMED)

# Define threshold
threshold = 0.8
loc = np.where( res >= threshold)

# Plot the found templates on the original image
for pt in zip(*loc[::-1]):
  cv2.rectangle(image, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

# Show image
plt.imshow(image)
plt.show()
```
