# Simple classifier

```python
# Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image
image1 = cv2.imread('../../data/square.png')
image2 = cv2.imread('../../data/triangle.png')

# Converting RGB image to grayscale
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Show both images
plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.show()
```

## Feature extraction

```python
# Compute edges
canny1 = cv2.Canny(gray1, 50, 100)
canny2 = cv2.Canny(gray2, 50, 100)

# Show both images
plt.subplot(1, 2, 1)
plt.imshow(canny1)
plt.subplot(1, 2, 2)
plt.imshow(canny2)
plt.show()
```

```python
# Compute Houghlines
lines1 = cv2.HoughLines(canny1, 10, np.pi*10/180, 100)
lines2 = cv2.HoughLines(canny2, 10, np.pi*10/180, 100)

# Print found lines for each image
print('The lines found in image one are: ' + str(lines1) + ', there are ' + str(len(lines1)) + ' lines\n')
print('The lines found in image two are: ' + str(lines2) + ', there are ' + str(len(lines2)) + ' lines')
```

## Classification

```python
# Classifier for each image
for lines in [lines1, lines2]:

    if len(lines) == 4:
        print("Square")
    elif len(lines) == 3:
        print("Triangle")
    else:
        print("Undefined")
```
